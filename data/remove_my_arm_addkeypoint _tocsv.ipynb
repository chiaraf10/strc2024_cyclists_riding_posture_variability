```python
folder_path = "key_points"  # Replace with the actual folder path
ann_path = "cyclist_dataset_annotations_STRC24.csv"

annotation = pd.read_csv(ann_path)
# create two olumn in the annotation dataframe to store the torso angle and arm angle
annotation["torso_angle"] = 0.0
annotation["arm_angle"] = 0.0

# Get a list of all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith(".csv")]

for file_name in annotation["object_id"]:
   try:
      #   Read the CSV file as a pandas dataframe
      files = [f for f in csv_files if f.startswith(file_name.split("_")[0]+"_object_id_"+ file_name.split("_")[1] )][0]
      
      kpts_csv = pd.read_csv(os.path.join(folder_path,files))
      # from the kpts_csv extract the row with the label "torso angle" and "arm angle"
      torso_angle = kpts_csv[kpts_csv["label"] == "torso_angle"]['x'].values[0]
      arm_angle = kpts_csv[kpts_csv["label"] == "arm_angle"]['x'].values[0]
      
      annotation.loc[annotation["object_id"] == file_name, "torso_angle"] = torso_angle
      annotation.loc[annotation["object_id"] == file_name, "arm_angle"] = arm_angle
   except:
      pass
```

```python
# remove the rows with the torso angle and arm angle equal to zero
annotation = annotation[annotation["torso_angle"] != 0]
```
